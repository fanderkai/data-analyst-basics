```python
# How many crime activities were recorded

%%sql
select
  count (*)
from
  CHICAGO_CRIME_DATA;
```

```python
# Which area with income per capita < 11000

%%sql
select
  community_area_number,
  community_area_name,
  per_capita_income
from
  CENSUS_DATA
where
  per_capita_income < 11000
```

```python
# Cases involving minors

%%sql
select
  id,
  case_number,
  primary_type,
  description
from
  CHICAGO_CRIME_DATA
where
  description like '%minor%'
```

```python
# Kidnapping cases involving children

%%sql
select
  id,
  case_number,
  primary_type,
  description
from
  CHICAGO_CRIME_DATA
where
  primary_type like '%kidnapping%'
  and
  description like '%child%'
```

```python
# Types of crimes that were commited at school

%%sql
select
  distinct primary_type,
  location_description
from
  CHICAGO_CRIME_DATA
where
  location_description like '%school%'
order by
  primary_type
```

```python
# Average safety score for each school types

%%sql
select
  "Elementary, Middle, or High School",
  avg(safety_score)
from
  CHICAGO_PUBLIC_SCHOOLS_DATA
group by
  "Elementary, Middle, or High School"
```

```python
# Top 5 area with highest % of households below poverty

%%sql
select
  community_area_name,
  max(percent_households_below_poverty) as highest_percent_households_below_poverty
from
  census_data
where
  percent_households_below_poverty is not null
group by
  community_area_name
order by
  highest_percent_households_below_poverty desc
limit 5
```

```python
# Top 1 crime number area

%%sql
select
  community_area_number,
  count(community_area_number) as CrimeNumber
from
  CHICAGO_CRIME_DATA
where
  community_area_number is not null
group by
  community_area_number
order by
  crimenumber desc
limit 1
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%%sql
select
  community_area_number,
  community_area_name,
  hardship_index
from
  census_data
where
  hardship_index = (
    select max(hardship_index)
    from census_data
  )
```

```python
# Top 1 crime number area

%%sql
select
  census.community_area_number,
  census.community_area_name,
  count(crime.community_area_number) as crimenumber
from
  CHICAGO_CRIME_DATA as crime
JOIN
  CENSUS_DATA as census
ON
  crime.community_area_number = census.community_area_number
WHERE
  census.community_area_name IS NOT NULL
GROUP BY
  census.community_area_number, census.community_area_name
ORDER BY
  CrimeNumber DESC
LIMIT 1;
```
