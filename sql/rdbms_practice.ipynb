```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect("university.db")
c = connection.cursor()
```

```python
# Create table to store data for faculties

c.execute('''CREATE TABLE faculties
              (id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
              faculty varchar(25)
              )''')
```

```python
# Add faculty instances

c.execute("INSERT INTO faculties (faculty) VALUES ('Arts & Sciences')")
c.execute("INSERT INTO faculties (faculty) VALUES ('Design')")
c.execute("INSERT INTO faculties (faculty) VALUES ('Law')")
c.execute("INSERT INTO faculties (faculty) VALUES ('Public Health')")
```

```python
# Check the faculties table

faculty_df = pd.read_sql_query("SELECT * FROM faculties", connection, index_col = 'id')
faculty_df
```

```python
# Create a departments table that has relational element to faculty through faculty_id as foreign key

c.execute('''CREATE TABLE IF NOT EXISTS departments (
              id INTEGER PRIMARY KEY AUTOINCREMENT,
              department varchar(25),
              faculty_id INTEGER,
              FOREIGN KEY (faculty_id) REFERENCES faculties(id)
          )''')
```
